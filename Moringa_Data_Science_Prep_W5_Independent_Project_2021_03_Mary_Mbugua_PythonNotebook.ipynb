#Let's read the data from the Autolib_dataset and create a dataframe

#1.1 DATA CLEANING

#(i)Missing data

```python
#We will now check how many missing vales we have in each column
Autolib_dataset.isnull().sum()
```

```python
#Dropping the columns with the missing values
Autolib_dataset.drop(["Displayed comment", "Scheduled at"], axis = 1, inplace = True)

```

```python
#Verifying that we no longer have null values in the entire dataframe
#and printing out the total number of missing values
print(Autolib_dataset.isnull().sum().sum())
```

# (ii).Duplicates

```python
#We will now identify which obseervations are duplicate
Autolib_dataset.duplicated()
```

```python
#Drop duplicate
Autolib_dataset.drop_duplicates().head(n=10)
```

#(iii)Converting the columns to lowercase

```python
Autolib_dataset.columns=map(str.lower,Autolib_dataset.columns)
Autolib_dataset
```

#(iv)Dropping unnecessary rows

```python
#Dropping rows that are unnecessary
Autolib_dataset.drop(['cars','charge slots','charging status','slots','geo point'],axis=1,inplace=True)
Autolib_dataset
```

#(vii)Replacing spaces in column names with (_)

```python
Autolib_dataset.columns = Autolib_dataset.columns.str.strip().str.replace(' ','_')
Autolib_dataset.columns
```

#(vi) JOINING OR MERGING COLUMNS

```python
from datetime import time
```

```python
#We are now going to convert hour and minute column from int to str
Autolib_dataset['hour']=Autolib_dataset['hour'].astype(str)
Autolib_dataset['minute']=Autolib_dataset['minute'].astype(str)

#Joining the two colums now
Autolib_dataset['Time']=Autolib_dataset['hour']+ ':' +Autolib_dataset['minute']
Autolib_dataset

```

```python
#Dropping the hour and minute columns
Autolib_dataset.drop(["hour", "minute"], axis = 1, inplace = True)
Autolib_dataset

```

```python
#Joining year,month and day columns to create a column named date
```

```python
Autolib_dataset['date']=pd.to_datetime((Autolib_dataset.year*10000+Autolib_dataset.month*100+Autolib_dataset.day).apply(str),format='%Y%m%d')
Autolib_dataset
```

#(viii) Saving the cleaned dataset

```python
#Converting the cleaned dataset to a csv file
Autolib_dataset.to_csv('Autolib.csv')
```
